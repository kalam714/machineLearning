```python
import numpy as np
import pandas as pd
import os, sys
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

```python
data=pd.read_excel('PahaW.xlsx')

data.shape
```

```python
data.head()
```

```python
label=data.loc[:,'Disease'].to_numpy()
from sklearn.preprocessing import LabelEncoder
labelencoder_y=LabelEncoder()
lebels=labelencoder_y.fit_transform(label)
lebels
```

```python
data2=data.drop(['Dominant hand'],axis=1)# dropping the status column
data2.head(10)
```

```python
features=data2[data2.columns[4:9]].to_numpy(dtype=object,na_value=0)
features
```

```python
scaler=MinMaxScaler((-1,1))
x=scaler.fit_transform(features)
y=lebels
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x, y, train_size=0.8, random_state=42)
print(X_train.shape)
print(Y_train.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

model_5 = GaussianNB()

model_5.fit(X_train,Y_train)

Y_pred = model_5.predict(X_test)

print(f1_score(Y_test, Y_pred)*100)
```
